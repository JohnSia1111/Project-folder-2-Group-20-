# Thermostaat

In het vorige werkblad heb je gezien hoe een gas zich gedraagt in een zuiger waarbinnen wordt voldaan aan de wet van behoud van energie.
In dat geval geldt $p \propto \alpha V^{-k}$, waarbij $k$ een exponent is die wordt bepaald door de vrijheidsgraden van het gas - in onze simulatie werken we met mono-atomaire gassen in 2D. 

We hadden kunnen denken dat we de ideale gaswet zouden kunnen gebruiken.
Echter, de temperatuur van het gas verandert door de werking van de zuiger - deze verricht arbeid. 

Het meenemen van arbeid dat verricht wordt door, of op, het gas is een eerste uitbreiding die belangrijk is in Thermodynamica.
Uitwisseling van energie (toevoegen of afvoeren van warmte) is een tweede uitbreiding.
In deze simulatie onderzoeken we die uitbreiding waarbij we warmte aan en af voeren uit het controlevolume.
Op die manier kunnen we ook isotherme simulaties doen.

- klasse voor het deeltje met bijbehorende functies

Daarna voegen we de code toe voor het warmtecontact:

- introduceren thermostaat

- bestuderen isotherm proces

`````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from scipy.optimize import curve_fit

BOX_SIZE_0 = 10              # Hoogte en lengte startvolume
N = 40                       # Aantal deeltjes
V_0 = 1                      # Startsnelheid van deeltjes
RADIUS = 0.5                   # Straal van moleculen
DT = 0.04                       # Tijdstap om geen botsing te missen
V_PISTON_0 = -0.1 * V_0      # Startsnelheid van zuiger 
k_B = 1.38E-23
# (negatief betekent zowel links als rechts naar binnen gericht)

`````

De klasse voor het gasmolecuul met de interacties:

`````python
class ParticleClass:
    def __init__(self, m, v, r, R):
        """ maakt een deeltje (constructor) """
        self.m = m                         
        self.v = np.array(v, dtype=float)  
        self.r = np.array(r, dtype=float)  
        self.R = R

    def update_position(self):
        """ verandert positie voor één tijdstap """
        self.r += self.v * DT 
            
    @property
    def momentum(self):
        return self.m * self.v
    
    @property
    def kin_energy(self):
        return 1/2 * self.m * np.dot(self.v, self.v)
    
def collide_detection(p1: ParticleClass, p2: ParticleClass) -> bool:
    """ Geeft TRUE als de deeltjes overlappen """
    return np.linalg.norm(p1.r - p2.r) < (p1.R + p2.R)


def particle_collision(p1: ParticleClass, p2: ParticleClass):
    """ past snelheden aan uitgaande van overlap """
    m1, m2 = p1.m, p2.m
    delta_r = p1.r - p2.r
    delta_v = p1.v - p2.v
    dot_product = np.dot(delta_r, delta_v)
    # Als deeltjes van elkaar weg bewegen dan geen botsing
    if dot_product >= 0: # '='-teken voorkomt ook problemen als delta_r == \vec{0}
        return
    distance_squared = np.dot(delta_r, delta_r) 
    # Botsing oplossen volgens elastische botsing in 2D
    p1.v -= 2 * m2 / (m1 + m2) * dot_product / distance_squared * delta_r
    p2.v += 2 * m1 / (m1 + m2) * dot_product / distance_squared * delta_r
`````

De randvoorwaarde van het volume. Hierbij is rekening gehouden met een bewegende zuiger die in het vorige werkblad is toegevoegd.

`````python
def top_down_collision(particle: ParticleClass):
    global impulse_outward, box_height
    if abs(particle.r[1]) + particle.R > box_height / 2:
        particle.r[1] = np.sign(particle.r[1]) * (box_height/2 - particle.R)
        impulse_outward += abs(particle.momentum[1]) * 2
        particle.v[1] *= -1

def left_right_collision(particle: ParticleClass):
    """ verzorgen van botsingen met wand links en rechts, die als zuiger kunnen bewegen """
    global box_length, v_piston, impulse_outward, work
    if abs(particle.r[0]) + particle.R > box_length / 2:
        particle.r[0] = np.sign(particle.r[0]) * (box_length/2 - particle.R)
        piston_velocity = np.sign(particle.r[0]) * v_piston
        relative_velocity = particle.v[0] - piston_velocity  # stelsel zuiger
        particle.v[0] = -relative_velocity + piston_velocity # stelsel waarnemer
        impulse_outward += 2 * particle.m * abs(relative_velocity)
        work += 2 * particle.m * relative_velocity * piston_velocity
`````

```{solution} ex-thermostat-02
In een 2D simulatie met deeltjes die dezelfde zijn, is de vrijheidsgraden gelijk aan 2. De gemiddele kinetische energie per deeltje is gelijk aan: $$\langle E_{kin} \rangle = \frac{f}{2} k_B T = k_B T$$

Dus de temperatuur wordt: $$T = \frac{E_{totaal}}{N \cdot k_B}$$

`````python
def create_particles(particles):
    """ Leegmaken en opnieuw aanmaken van deeltjes  in lijst """
    global box_length, box_height
    particles.clear()
    for _ in range(N):
        vx = np.random.uniform(-V_0, V_0)
        vy = np.random.choice([-1, 1]) * np.sqrt(V_0**2 - vx**2)        
        x = np.random.uniform(-box_length/2 + RADIUS, box_length/2 - RADIUS)
        y = np.random.uniform(-box_height/2 + RADIUS, box_height/2 - RADIUS)
        particles.append(ParticleClass(m=1.0, v=[vx, vy], r=[x, y], R=RADIUS))
    
def temperature(particles) -> float:
    total_kin_energy = sum(p.kin_energy for p in particles)
    # In 2D is E_avg = k_B * T (want f=2, dus 2/2 * k_B * T)
    temp = total_kin_energy / (len(particles) * k_B)


    return temp
        
def handle_collisions(particles):
    """ alle onderlinge botsingen afhandelen voor deeltjes in lijst """
    num_particles = len(particles)
    for i in range(num_particles):
        for j in range(i+1, num_particles):
            if collide_detection(particles[i], particles[j]):
                particle_collision(particles[i], particles[j])

def handle_walls(particles):
    """ botsing met wanden controleren voor alle deeltjes in lijst en gemiddeld bepaling druk """
    global pressure, impulse_outward, box_height, box_length   # om pressure buiten de functie te kunnen gebruiken
    impulse_outward = 0.0
    for p in particles:
        left_right_collision(p)
        top_down_collision(p)    
    pressure = 0.95 * pressure + 0.05 * impulse_outward / ((2 * box_length + 2 * box_height) * DT) 
    
def take_time_step(particles):
    """ zet tijdstap voor een lijst deeltjes en verwerk alle botsingen onderling en met wanden """
    global box_length, v_piston
    box_length += 2 * v_piston * DT # zowel links als rechts zuiger
    for p in particles:
        p.update_position()
    handle_collisions(particles)
    handle_walls(particles)  
`````

## De thermostaat

In de werkelijkheid is de temperatuur van de wand een mate van de amplitude van de trillingen van de deeltjes waaruit de wand bestaat.
De wederzijdse overdracht van de energie van die trillingen naar de kinetische energie van de gasmoleculen bepaalt het thermische contact tussen de wanden van het volume en het gas.
In ons model bestaat de wand echter helemaal niet uit deeltjes maar hebben we een denkbeeldige lijn getrokken in de ruimte.
We moeten daarom een wiskundige truc toepassen om de temperatuur van het gas te beïnvloeden. 

Er zijn in de literatuur verschillende van dit soort trucs bedacht.
Ze worden een 'thermostaat' genoemd.
Elk van die type trucs hebben hun voor- en nadelen.
In ons geval houden we het simpel: Op het moment dat een gasmolecuul botst met de wand (die een thermisch contact voorstelt), dan schalen we de snelheid van dit molecuul met (de wortel van) de verhouding tussen de veronderstelde temperatuur van de wand en de temperatuur die het gas op dat moment heeft:

$$
  v = \sqrt{ \frac{T_{wand}} {T_{deeltje}} }v
$$

Voor het gemak houden we de linker en rechter wand van het volume als zuiger en maken we het thermische contact aan de onder- en bovenwand.

`````python
def top_down_collision(particle: ParticleClass) -> None:
    """ verzorgen van botsingen met wand boven en onder, die als thermostaat kunnen werken """
    global box_height, set_temp, impulse_outward, heat
    if abs(particle.r[1]) + particle.R > box_height / 2:
        temp_factor = (set_temp/temperature(particles)) if set_temp > 0 else 1.0  # Verhouding tussen de gewenste energie en de huidge energie van het gas.
        particle.r[1] = np.sign(particle.r[1]) * (box_height/2 - particle.R) # tunneliung voorkomen
        impulse_outward += abs(particle.momentum[1]) * (1 + temp_factor**0.5) # Impulsoverdracht
        heat += particle.kin_energy * (temp_factor - 1) # Toevoegen van warmte
        particle.v *= temp_factor**0.5 #Thermische interactie
        particle.v[1] *= -1 #De richting loodrecht op de wand
`````

```{solution} ex-thermostat-05
De temperatuur van het gas wordt bepaald door de gemiddelde kinetische energie van alle deeltjes. De "thermostaat" werkt echter alleen aan de randen. Wanneer de zuiger beweegt, wordt er arbeid verricht op de deeltjes die tegen de zuiger botsen. Deze extra energie (of energieverlies bij expansie) moet zich eerst via onderlinge botsingen verspreiden door het gas (diffusie van energie) voordat de deeltjes de thermostaatwanden bereiken om de energie weer af te staan. Hierdoor is er sprake van een relaxatietijd. Omdat de zuiger continu beweegt, is het systeem nooit in perfect thermodynamisch evenwicht met de wand; het loopt altijd iets achter de feiten aan.
```

```{solution} ex-thermostat-06
Bij een klein volume is de afstand tussen de zuigerwanden klein. De frequentie waarmee deeltjes tegen de bewegende zuiger botsen neemt daardoor drastisch toe ($f_{botsing} \propto 1/L$). Hierdoor neemt de hoeveelheid arbeid die per tijdseenheid op het gas wordt verricht toe ($P_{in} = \frac{dW}{dt}$).


De thermostaat bevindt zich aan de boven- en onderwand. Als de verhouding oppervlakte/volume verandert, of simpelweg als de energie-instroom door arbeid groter wordt dan de maximale energie-uitstroom (warmteafvoer) via de thermostaatwanden, hoopt de energie zich op. Het systeem wordt dan effectief adiabatisch in plaats van isotherm, waardoor $T$ stijgt.
```

### Informatie
De Eerste Hoofdwet van de Thermodynamica luidt:$$\Delta U = Q + W$$

Waarbij:

$\Delta U$: Verandering in interne energie ($E_{kin, eind} - E_{kin, start}$).


$Q$: Warmte toegevoegd aan het systeem (heats array).


$W$: Arbeid verricht op het systeem (works array).

Om dit te verifiëren is de verandering in interne energie geplot. Dit is verder vergeleken met de som van arbeid en warmte.

In het boek wordt uitgelegd dat er maar twee vormen van energieoverdracht zijn van een systeem naar de omgeving. Dit kan via warmteoverdracht of via arbeid.
Bijzonder is dat deze beide grootheden alleen de snelheid van de moleculen beïnvloeden.
Toch zullen we in het volgende werkblad zien dat er een fundamenteel verschil zit in de werking van deze twee macroscopische grootheden.

- Geef de warmtegeleidingscoëfficiënt in W/K van het ideaalgas in je simulatie. (Opmerking: De warmtegeleidingscoëfficiënt wordt normaal gegeven in W/mK, maar in onze tweedimensionale simulatie valt de lengteafhankelijkheid er uit)

`````python
# De nieuwe opstelling
BOX_HEIGHT_EXT = 4 * BOX_SIZE_0    # 4x zo hoog
BOX_LENGTH_EXT = BOX_SIZE_0        # Zelfde breedte
N_EXT = 160                        # 4x zoveel deeltjes voor constante dichtheid
V_PISTON_EXT = 0                   # Stilstaande zuiger

# Variabelen voor warmtestroom
heat_flow_top = 0.0
heat_flow_bottom = 0.0

#  Aangepaste collision functie
def top_down_collision_conduction(particle):
    global heat_flow_top, heat_flow_bottom, T_top, T_bot, box_height
    
    # botsing met boven- of onderwand
    if abs(particle.r[1]) + particle.R > box_height / 2:
        
        # Bepaal welke wand en welke temperatuur
        if particle.r[1] > 0:
            # Bovenwand (Warm)
            target_temp = T_top
            is_top = True
        else:
            # Onderwand (Koud)
            target_temp = T_bot
            is_top = False
            
        # Huidige temperatuur van het deeltje (benaderd via kinetische energie)
        # E_kin = k_B * T_part -> T_part = E_kin / k_B
        # Maar de scaling factor werkt op basis van bulk T, of individueel?
        # De originele code gebruikt de bulk temperatuur 'temperature(particles)'. 
        # Voor correcte thermalisatie gebruiken we hier de bulk T van het gas dichtbij de wand,
        # maar eenvoudigheidshalve gebruiken we de globale T van het gas zoals in de originele thermostaat.
        
        current_gas_temp = temperature(particles) 
        temp_factor = (target_temp / current_gas_temp)
        
        # Positie correctie
        particle.r[1] = np.sign(particle.r[1]) * (box_height/2 - particle.R)
        
        # Energieboekhouding (dQ)
        # dQ > 0 betekent warmte NAAR het gas
        dE = particle.kin_energy * (temp_factor - 1)
        
        if is_top:
            heat_flow_top += dE
        else:
            heat_flow_bottom += dE
            
        # Snelheid aanpassen
        particle.v *= temp_factor**0.5
        particle.v[1] *= -1

# 3. Setup van de simulatie
particles = []
# Overschrijf globale variabelen voor de functies die ze gebruiken
box_height = BOX_HEIGHT_EXT
box_length = BOX_LENGTH_EXT
# Let op: create_particles gebruikt de globale box_height/length
create_particles(particles) 

# Breid aantal deeltjes uit naar 160 (create_particles maakt er N aan, dus we roepen hem vaker aan of passen N tijdelijk aan)
# Makkelijkste: particles lijst opnieuw vullen met 160 deeltjes handmatig of N aanpassen
particles.clear()
for _ in range(N_EXT):
        vx = np.random.uniform(-V_0, V_0)
        vy = np.random.choice([-1, 1]) * np.sqrt(V_0**2 - vx**2)        
        x = np.random.uniform(-box_length/2 + RADIUS, box_length/2 - RADIUS)
        y = np.random.uniform(-box_height/2 + RADIUS, box_height/2 - RADIUS)
        particles.append(ParticleClass(m=1.0, v=[vx, vy], r=[x, y], R=RADIUS))

# Temperaturen instellen
T_start = temperature(particles)
T_top = 1.5 * T_start
T_bot = 0.5 * T_start

heat_flow_top = 0.0
heat_flow_bottom = 0.0

# Arrays voor data
steps = 2000
time_array = np.arange(steps) * DT
heat_top_cum = np.zeros(steps)

# Simulatie loop
print("Start simulatie warmtegeleiding...")
for i in range(steps):
    # Stap nemen (gebruik aangepaste botsing logic)
    for p in particles:
        p.update_position()
    
    # Botsingen onderling
    handle_collisions(particles)
    
    # Botsingen wand (hier roepen we onze NIEUWE functie aan i.p.v. handle_walls compleet)
    # We doen left_right (zuiger stil) en top_down (thermostaat) apart
    for p in particles:
        # Link/Rechts is gewoon elastisch want V_piston = 0
        if abs(p.r[0]) + p.R > box_length / 2:
            p.r[0] = np.sign(p.r[0]) * (box_length/2 - p.R)
            p.v[0] *= -1
        
        # Boven/Onder met de nieuwe functie
        top_down_collision_conduction(p)
        
    heat_top_cum[i] = heat_flow_top

# Plotten van de warmtestroom
plt.plot(time_array, heat_top_cum)
plt.xlabel('Tijd [s]')
plt.ylabel('Toegevoegde Warmte Q_top [J]')
plt.title('Cumulatieve warmte instroom via bovenwand')
plt.show()
`````

### Analyse
Voor de warmtestroom geldt:
$$P = \frac{dQ}{dt} = -\kappa \cdot A \cdot \frac{dT}{dy}$$

In onze discretisatie benaderen we dit als:$$P \approx \kappa \cdot L_x \cdot \frac{T_{top} - T_{bot}}{L_y}$$
$L_x$  wordt $A$.

### Stappenplan voor de berekening:

Bepaal $P$: De helling van de grafiek heat_top_cum vs time is de warmtestroom $P$ (in Joule per tijdstap-eenheid, let op de $dt$).Neem het lineaire deel van de grafiek (zodra steady-state is bereikt).$P = \frac{\Delta Q}{\Delta t}$.Bereken $\kappa$: Herschrijf de formule:$$\kappa = P \cdot \frac{L_y}{L_x \cdot (T_{top} - T_{bot})}$$
